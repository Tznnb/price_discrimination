```python

import numpy as np
import sys, os
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

from algorithms.market import Market
from experiments.feasibility_grid import run_feasibility_grid_experiment
from utils.visualization import plot_feasibility_heatmap
```

```python

def generate_uniform_distribution(n=5, low=1, high=10):
    values = np.linspace(low, high, n, dtype=int)
    masses = np.ones_like(values) / n
    return values.tolist(), masses.tolist()

def generate_binary_distribution(low=1, high=10, p=0.3):
    values = [low, high]
    masses = [p, 1 - p]
    return values, masses

def generate_geometric_distribution(start=1, ratio=2, n=5, q=0.5):
    values = [start * (ratio ** i) for i in range(n)]
    masses = [q ** i for i in range(n)]
    masses = masses / np.sum(masses)
    return values, masses.tolist()
```

## üîπ Part AÔºöÂü∫Á°ÄÂàÜÂ∏É‰∏ãÁöÑÂèØË°åÊÄßÁÉ≠Âõæ

```python

distributions = {
    "Uniform_5": generate_uniform_distribution(5),
    "Binary_03": generate_binary_distribution(p=0.3),
    "Binary_05": generate_binary_distribution(p=0.5),
    "Binary_07": generate_binary_distribution(p=0.7),
    "Geometric_q3": generate_geometric_distribution(q=0.3),
    "Geometric_q5": generate_geometric_distribution(q=0.5),
    "Geometric_q7": generate_geometric_distribution(q=0.7),
}

for name, (vals, masses) in distributions.items():
    print(f"--- {name} ---")
    print(f"values = {vals}")
    print(f"masses = {np.round(masses, 3)}\n")
    run_feasibility_grid_experiment(vals, masses, debug=False)
```
