```python

import numpy as np
import sys, os
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

from algorithms.market import Market
from experiments.feasibility_grid import generate_feasibility_grid, run_feasibility_grid_experiment
from utils.visualization import plot_feasibility_heatmap
from utils.market_generator import MarketGenerator
from utils.feasibility_analysis import analyze_feasibility_matrix
```

## üîπ Part AÔºöÂü∫Á°ÄÂàÜÂ∏É

```python

part_a_distributions = {
    "Uniform_5": MarketGenerator.uniform(n=5),
    "Binary_03": MarketGenerator.binary(p=0.3),
    "Binary_05": MarketGenerator.binary(p=0.5),
    "Geometric_q5": MarketGenerator.geometric(q=0.5),
}

for name, (vals, masses) in part_a_distributions.items():
    print(f"\n--- {name} ---")
    print(f"values = {vals}")
    print(f"masses = {np.round(masses, 3)}")
    market = Market(vals, masses)
    matrix = generate_feasibility_grid(market)
    plot_feasibility_heatmap(vals, matrix)
    analyze_feasibility_matrix(vals, matrix)
```

## üîπ Part BÔºöÂ§çÊùÇÂàÜÂ∏É

```python

part_b_distributions = {
    "Normal_mu5_sigma1": MarketGenerator.truncated_normal(mu=5, sigma=1, n=5),
    "Normal_mu5_sigma3": MarketGenerator.truncated_normal(mu=5, sigma=3, n=5),
    "Bimodal_even": MarketGenerator.bimodal([1, 2], [9, 10], weight=0.5),
    "PowerLaw_a2": MarketGenerator.powerlaw(alpha=2.0, n=5),
}

for name, (vals, masses) in part_b_distributions.items():
    print(f"\n--- {name} ---")
    print(f"values = {vals}")
    print(f"masses = {np.round(masses, 3)}")
    market = Market(vals, masses)
    matrix = generate_feasibility_grid(market)
    plot_feasibility_heatmap(vals, matrix)
    analyze_feasibility_matrix(vals, matrix)
```
